```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score # cross_val_score ekledim
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
import math
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model # bunu ekledim
from sklearn import decomposition, datasets # bunu ekledim
```

```python
data_full = pd.read_csv("A.csv")
data = data_full[["sales", "price", "basket", "fav", "visit", "impression", "stock", "demand", "removeFromFav", "reviewCount", "rating"]]
data
```

```python
#number of product to be regressed
n = 250
sample_data = data[0:207*n]
```

```python
#FIT LASSO REGRESSION MODEL
#define predictor and response variables
X = sample_data[["price", "basket", "fav", "visit", "impression", "stock", "demand", "removeFromFav", "reviewCount", "rating"]]
y = sample_data["sales"]
X_train = X[0:207*math.floor(4*n/5)]
X_test = X[207*math.floor(4*n/5):207*n]
y_train = y[0:207*math.floor(4*n/5)]
y_test = y[207*math.floor(4*n/5):207*n]
```

```python
pipeline = Pipeline([
                     ('scaler',StandardScaler()),
                     ('model',Lasso())
])
```

```python
#Searching for best alpha
search = GridSearchCV(pipeline,
                      {'model__alpha':np.arange(1e-7,1e-5,1e-7)},
                      cv = 5, scoring="neg_mean_squared_error",verbose=3
                      )
```

```python
#fit model
search.fit(X_train,y_train)
```

```python
coefficients = search.best_estimator_.named_steps['model'].coef_
search.best_estimator_
```

```python
alph= 1e-7
lasso= Lasso(alpha=alph)
lasso.fit(X_train, y_train)
y_train_pred= lasso.predict(X_train)
y_test_pred= lasso.predict(X_test)
print(np.floor(y_train_pred))
print(y_train)
print('train R2 score is',mean_absolute_error(y_train,y_train_pred))
print('test R2 score is',mean_absolute_error(y_test,y_test_pred))
```

```python
importance = np.abs(coefficients)
importance
```

```python
#from here the elasticNet implementation starts
pipe = Pipeline(steps=[('stc_slc', StandardScaler()),
                           ('pca', PCA()),
                           ('elasticnet', ElasticNet())])
```

```python
n_components = list(range(1,X.shape[1]+1,1))
normalize = [True, False]
selection = ['cyclic', 'random']
```

```python
parameters = dict(pca__n_components=n_components,
                      elasticnet__normalize=normalize,
                          elasticnet__selection=selection)
```

```python
clf_EN = GridSearchCV(pipe, parameters)
clf_EN.fit(X, y)
```

```python
print('Best Number Of Components:', clf_EN.best_estimator_.get_params()['pca__n_components'])
print(clf_EN.best_estimator_.get_params()['elasticnet'])
```
