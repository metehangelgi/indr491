```python
#MAPE function
def mape(actual, pred):
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs((actual-pred)/ actual))*100
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#slice data with complete info (5 periods)
Dates =df.iloc[5:,0]
Price = df.iloc[5: 1]
MovingAverage=df.iloc[5:,2]
```

```python
#report MSE
print("MSE: " , mean_squared_error(Price, MovingAverage))
#report MAPE
print("MAPE: " , round(mape(Price, MovingAverage),2))
```
